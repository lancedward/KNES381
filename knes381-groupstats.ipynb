```python
# import numpy as np # Commented out becuase we imported this in the previous cell
# create a group of people that matach the stats can standard canadian
# Men Height 176.5 cm with a standard deviation of 1.2 cm
# Women height 162.1 with a standard deviation of 1.5 cm
np.random.seed(0)
group_height = np.random.normal(loc=169,scale=7,size=25)
# to round this to two decimal places you would round the outside information
# group_height = np.round(np.random.normal(loc=169,scale=7,size=25),2)
print(group_height, '\n')
print(group_height.std(),group_height.mean()) # 5.62761460277423 10.702796361565493
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

# Plot Histogram on group_height
plt.hist(group_height, bins=5)
plt.gca().set(title='Frequency Histogram Height', ylabel='Frequency', xlabel='Height (cm)')
np.savetxt('group_height.txt', group_height)
```

```python
np.random.seed(0)
group_weight = np.random.normal(loc=73.5,scale=15.7,size=25)
print(group_weight)
print(group_weight.std(),group_weight.mean()) # 5.62761460277423 10.702796361565493
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

# Plot Histogram on group_height
plt.hist(group_weight, bins=5)
plt.gca().set(title='Frequency Histogram weight', ylabel='Frequency', xlabel='Weight (kg)')
np.savetxt('group_weight.txt', group_weight)
```
